## `3.` Create another notebook to answer some questions using SQL Language

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('scraped_DB/book_scrapping.db')
```

### `A`-write a sql query to get books that has rate >=3 and has 'Mr' in its name.

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.rate      AS 'Book Rate' 

FROM 
_0_1_books  AS b 

WHERE b.rate>=3 AND b.book_name LIKE '%Mr%' 
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `B`-write a sql query to get books that has rate >=3 or price > 20.

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.rate      AS 'Book Rate',
b.price     AS 'Book Price' 

FROM 
_0_1_books  AS b 

WHERE b.rate>=3 OR b.price>20
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `C`-write a sql query to get books that has rate not 3.

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.rate      AS 'Book Rate'

FROM 
_0_1_books  AS b 

WHERE b.rate!=3
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `D`-write a sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.rate      AS 'Book Rate',
b.price     AS 'Book Price'

FROM 
_0_1_books  AS b 

WHERE (b.price BETWEEN 10 AND 40) AND b.rate IN (1,3,4)
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `E`-write a sql query to get the top 5 most expensive books (order books desc with price column and then limit the result to the first 5).

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.price     AS 'Book Price'

FROM 
_0_1_books  AS b 

ORDER BY b.price DESC LIMIT 5
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `F`-write a sql query to get the 3rd 10 books in the books table order first by rate desc and then by price asc.

```python
SQL_Query = """
SELECT 

b.book_id   AS 'Book ID'  , 
b.book_name AS 'Book Name',
b.rate      AS 'Book Rate',
b.price     AS 'Book Price'

FROM 
_0_1_books  AS b 

ORDER BY b.rate DESC, b.price ASC LIMIT 10
"""
df = pd.read_sql(SQL_Query, conn)
df
```

```python
def insertCategory(category_name="",connection=""):
        SQL = """
            INSERT INTO "main"."_0_0_categories"
            ("cat_name")
            VALUES (?);
            """
        Data = [category_name]
        row = connection.execute(SQL,Data)
        connection.commit()
        return row.lastrowid
    
def insertBook(book_name="",price="",rate="",available_quantity="",cat_id="",connection=""):
    SQL = """
        INSERT INTO "main"."_0_1_books"
        (book_name,price,rate,available_quantity,cat_id)
        VALUES (?,?,?,?,?);
        """
    Data = [book_name,price,rate,available_quantity,cat_id]
    row = connection.execute(SQL,Data)
    connection.commit()
    return row.lastrowid
```

### `N`-write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.

```python
SQL_Query = f"""
SELECT 
 
b.book_name AS 'Book Name',
c.cat_name  AS 'Category Name',
b.rate      AS 'Book Rate',
b.price     AS 'Book Price'

FROM 
_0_0_categories AS c
INNER JOIN
_0_1_books      AS b 

WHERE c.cat_id = b.cat_id
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `O`-write a sql query to calculate how many books each rate has and have price between 20£ and 30£.

```python
SQL_Query = f"""
SELECT 

COUNT(*) AS 'Number Of Books'

FROM _0_1_books      AS b 

WHERE b.price BETWEEN 20 AND 30
"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `P`-write a sql query to calculate how many books each category has having count > 10.

```python
SQL_Query = """
SELECT 

b.cat_id   AS 'Category ID', 
c.cat_name AS 'Category Name', 
COUNT(*)   AS 'Number Of Books'

FROM 

_0_0_categories AS c
INNER JOIN
_0_1_books      AS b

WHERE c.cat_id = b.cat_id

GROUP BY b.cat_id

HAVING COUNT(*) > 10

ORDER BY COUNT(*)

"""
df = pd.read_sql(SQL_Query, conn)
df
```

### `Q`-write a sql query to get all books with category_name='Music' using subquery.

```python
SubQuerySelect = """
SELECT cat_id FROM _0_0_categories
WHERE cat_name = 'Music'
"""

SQL_Query = f"""
SELECT 

b.book_name AS 'Book Name',
c.cat_name  AS 'Category Name',
b.rate      AS 'Book Rate',
b.price     AS 'Book Price'

FROM 

_0_0_categories AS c
INNER JOIN
_0_1_books      AS b

WHERE c.cat_id = b.cat_id
AND b.cat_id IN ({SubQuerySelect});

"""
df = pd.read_sql(SQL_Query, conn)
df
```

## `4.` Export Database to CSV File and use pandas to read this CSV.

#### `A.` Select All Data From DataBase And Generate CSV File

```python
SQL_Query = """
SELECT 

c.cat_id                 AS 'Category ID',
c.cat_name               AS 'Category Name',
b.book_id                AS 'Book ID',
b.book_name              AS 'Book Name',
b.rate                   AS 'Book Rate',
b.price                  AS 'Book Price',
b.available_quantity     AS 'Book Quantity'

FROM 

_0_0_categories AS c
INNER JOIN
_0_1_books      AS b

WHERE c.cat_id = b.cat_id

"""
df = pd.read_sql(SQL_Query, conn)
df.to_csv('scraped_DB/book_scrapping.csv')
print('Data Converted To CSV File.....')
```
