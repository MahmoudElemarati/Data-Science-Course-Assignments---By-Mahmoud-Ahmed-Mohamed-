```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('scraped_DB/book_scrapping.db')
```

## Get Repeated Books In the Site

```python
SQL_Query = f"""
SELECT 

b.book_name        AS 'Book Name',
c.cat_name         AS 'Category Name',
COUNT(b.book_name) AS 'Book Count'

FROM 

_0_0_categories AS c
INNER JOIN
_0_1_books      AS b

WHERE c.cat_id = b.cat_id
GROUP BY b.book_name
HAVING COUNT(b.book_name)>1;

"""
df = pd.read_sql(SQL_Query, conn)
df
```
